```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips = pd.read_csv("../tips.csv")
```

```python
palette = sns.color_palette(
            "hls", len(np.unique(tips["size"])))
```

```python
sns.kdeplot(x="total_bill", data=tips, hue="size", common_norm=False, alpha=.3, linewidth=1, fill=True, palette=palette)
```

```python
plt.subplots(figsize=(10, 10))
sns.set_theme(style='darkgrid')  # 图形主题
sns.kdeplot(x="total_bill", data=tips, common_norm=False, alpha=.3, linewidth=1, fill=True, palette=palette)
tips["total_bill2"] = tips["total_bill"] + np.random.rand(tips.shape[0]) + 10
sns.kdeplot(x="total_bill2", data=tips, common_norm=False, alpha=.3, linewidth=1, fill=True, palette=palette)
plt.legend(labels=["1", "2"])
```
